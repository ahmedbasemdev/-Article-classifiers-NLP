```python
import pandas as pd 
import numpy as np 
import os
import string
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
path = os.path.normpath(os.getcwd() + os.sep + os.pardir) + "\[1] Data Cleaning" + "\_finalCleaned.csv"
print(path)
```

```python
data = pd.read_csv(path).drop('Unnamed: 0',axis=1)
```

## Data Preprocessing 

```python
punctuation = string.punctuation
```

```python
def removePunctuation(text):
    text = "".join([word for word in text if word not in punctuation])
    return text
```

```python
print(removePunctuation('hello !!'))
```

```python
data['text'] = data['Article Title'].apply(removePunctuation)
```

```python
data.head()
```

```python
data['textLower'] = data['text'].apply(lambda x :x.lower())
```

```python
data.head()
```

```python
data['Tokenized'] = data['textLower'].apply(word_tokenize)
```

```python
data.head()
```

```python
stopWords = nltk.corpus.stopwords.words('english')
```

```python
def removeStopWords(text):
    text= [word for word in text if word not in stopWords]
    return text
```

```python
data['NoStopWords'] = data['Tokenized'].apply(removeStopWords)
```

```python
data.head()
```

```python
porter_stemmer = PorterStemmer()
def stemmingText(text):
    text = [porter_stemmer.stem(word) for word in text]
    return text
```

```python
data['stemmedText'] = data['NoStopWords'].apply(stemmingText)
```

```python
data.head()
```

```python
wordLemmatizer = WordNetLemmatizer()
def lemmatizingText(text):
    text = [wordLemmatizer.lemmatize(word) for word in text]
    return text
```

```python
data['lemmtizedText'] = data['stemmedText'].apply(lemmatizingText)
```

```python
data['finalText'] = data['stemmedText'].apply(lambda x :" ".join(x))
```

```python
data.head()
```

```python
data = data[['finalText','Category']]
```

```python
data.head()
```
